## Part 5: Working with 2D Lists (Arrays)

### Introduction to 2D Lists

A 2D list is a list of lists - like a grid or table.

````python
# Creating a 2D list (3x3 grid)
grid = [
    ['.', '.', '.'],
    ['.', 'X', '.'],
    ['.', '.', '.']
]

# Printing the grid
for row in grid:
    for cell in row:
        print(cell, end=' ')
    print()  # New line after each row
````

````python
two_d_list = [[1,2,3],[4,5,6],[7,8,9]]
#first argument is which list
#second argument is within that, which number
print(two_d_list[0][2])

#which 2d list
#which 1d list
#which number
three_d_list = [[[1,2,3],[4,5,6],[7,8,9]],[[.1,.2,.3],[.4,.5,.6],[.7,.8,.9]],[[.01,.02,.03],[.04,.05,.06],[.07,.08,.09]]]
print(three_d_list[2][2][1])

#possible to have lists of lists - one within the other
#2d list is just a list of 2d lists
````

````python
# Accessing elements in a 2D list
grid = [
    ['A', 'B', 'C'],
    ['D', 'E', 'F'],
    ['G', 'H', 'I']
]

print("Element at row 0, col 0:", grid[0][0])  # 'A'
print("Element at row 1, col 1:", grid[1][1])  # 'E'
print("Element at row 2, col 2:", grid[2][2])  # 'I'

# Modifying elements
grid[1][1] = 'X'
print("\nAfter modification:")
for row in grid:
    print(row)
````

### Creating a 2D List with Loops

````python
# Create a 5x5 grid filled with '.'
SIZE = 5
grid = []

for row in range(SIZE):
    row_list = []
    for col in range(SIZE):
        row_list.append('.')
    grid.append(row_list)

# Print the grid
for row in grid:
    for cell in row:
        print(cell, end=' ')
    print()
````

````python
# Alternative: Using list comprehension
SIZE = 5
grid = [['.' for col in range(SIZE)] for row in range(SIZE)]

# Print the grid
for row in grid:
    print(' '.join(row))
````

### Exercise 7: Simple Grid Game

Create a simple grid-based game where a player ('P') can move around.

**Requirements:**
1. Create an 8x8 grid filled with '.'
2. Place a player 'P' at position (4, 4)
3. Print the grid
4. Ask the user for a direction (u/d/l/r for up/down/left/right)
5. Move the player (mark old position with '-')
6. Print the updated grid
7. Repeat until the user types 'q' to quit

**Example:**
```
. . . . . . . .
. . . . . . . .
. . . . . . . .
. . . . . . . .
. . . . P . . .
. . . . . . . .
. . . . . . . .
. . . . . . . .

Move (u/d/l/r/q): r

. . . . . . . .
. . . . . . . .
. . . . . . . .
. . . . . . . .
. . . . - P . .
. . . . . . . .
. . . . . . . .
. . . . . . . .
```

**Hints:**
- Keep track of player position with two variables: `player_row` and `player_col`
- Before moving, set the current position to '-'
- Update the row/col based on direction
- Set the new position to 'P'

````python
# Exercise 7: Simple Grid Game
# Write your code here

from IPython.display import clear_output

SIZE = 8 #defining the grid size 
grid = [] #will hold the entire grid (list of rows)

#Create 8x8 grid filled with
for row in range(SIZE): #looping 8 times to create 8 rows
    row_list = [] #everytime theres a new row, you start with an empty list
    for col in range (SIZE): #inner loop that runs 8 times
        row_list.append('.') #this loop creates a row with 8 dots 
    grid.append(row_list) #adds that row to your grid

#player starting position
player_row = 4
player_col = 4
grid[player_row][player_col] = 'P' #replaces the dot with P

#Game Loop
while True: #creates an infinite loop until exited 
    clear_output(wait=True) #clears screen each time the loop runs

    #Print the grid
    for row in grid: #looping through each row in the grid
        for cell in row: #loop through each cell inside that row
            print(cell,end="") #prints without starting a new line
        print() #prints a blank line which moves you to the next row

    move = input("Move (u/d/l/r/q): ") #asks user for a direction

    if move == 'q': #break if user enters q
        break

    #Mark old position
    grid[player_row][player_col] = '-' #before moving the player, mark the old spot with a dash

    #Update position
    if move == 'u' and player_row > 0: #checks did user type up, and not already at the top of grid
        player_row -=1 #go to row above (gets smaller)
    elif move == 'd' and player_row < SIZE - 1: #checks did user type down, and not already at the bottom of grid
        player_row += 1 #go to row below (gets bigger)
    elif move == 'l' and player_col > 0: #checks did user type left, and not already at the left of grid
        player_col -= 1 #move to left, gets smaller
    elif move == 'r' and player_col <  SIZE - 1: #checks did user type right, and not already at the right of grid
        player_col += 1 #move to right, gets bigger

    #Set new position
    grid[player_row][player_col] = 'P' #replace new position with P

#join function is more efficient
````

## Summary: Key Concepts

### String Operations:
- **Concatenation**: `+` operator or f-strings
- **Formatting**: f-strings with format specifiers `{value:.2f}`
- **Splitting**: `.split()` method
- **Searching**: `in` keyword, `.find()` method

### String Methods:
- `.upper()`, `.lower()` - Change case
- `.strip()` - Remove whitespace
- `.replace()` - Replace substrings
- `.startswith()`, `.endswith()` - Check beginnings/endings
- `.isalpha()`, `.isdigit()`, `.isalnum()` - Character classification

### String Indexing:
- Access characters: `string[index]`
- Length: `len(string)`
- Reverse: `string[::-1]`
- Slicing: `string[start:end]`

### 2D Lists:
- Access: `grid[row][col]`
- Iterate with nested loops
- Create with list comprehension
