## If, elif, else statements example: Roller Coaster Height Checker

Before we start the exercises, let's look at an example program that determines whether someone is tall enough to ride a roller coaster.

**Requirements:**
- Get the user's height
- If height is 0 or less, print an error message
- If height is below minimum (100cm), tell them they cannot ride
- If height is between 100cm and 160cm, tell them they can ride with an adult
- If height is 160cm or above, tell them they can ride alone

**Example solution:**

````python
#Rollercoaster Height Check
MIN_HEIGHT = 100
RIDE_ALONE_HEIGHT = 160
height = int(input("enter your height in cm"))
if height <= 0:
    print("Error! Invalid height")
elif height<MIN_HEIGHT:
    print("Sorry, you are not tall enough to ride.")
elif height< RIDE_ALONE_HEIGHT:
    print("You can ride with an adult.")
else:
    print("You can ride alone.")


````

## Exercise 1: Addition

Create a program that adds together the number of students and tutors in a class.

**Requirements:**
- Ask for two integers with the message: `"Please enter the number of students and tutors: "`
- Display the sum in the format: `students + tutors = sum`

**Examples:**
```
Please enter the number of students and tutors: 2 5
2 + 5 = 7
```

```
Please enter the number of students and tutors: -1 5
-1 + 5 = 4
```

**Hint:** Use `input().split()` to read two numbers on one line, then convert them to integers.

````python
# Exercise 1: Addition
# Write your code here

students_and_tutors = input("Please enter number of students and number of tutors.")
print(students_and_tutors)
students, tutors = students_and_tutors.split()
print(students)
print(tutors)

total_people = int(students)+int(tutors)
print(total_people)




````
